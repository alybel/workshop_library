# A typical data analysis workflow

## Import Statements

```python
# External libraries
import pandas as pd # pandas library for tabular data calculations
%matplotlib inline
```

## Load data and prepare

```python
# read csv data into the jupyter notebook environment
df = pd.read_csv('../../intraday_1min_MSFT.csv', parse_dates=['timestamp'])
```

```python
df.head()
```

```python
df['trading_hour'] = df['timestamp'].dt.hour
```

```python
df.head()
```

```python
df = df.sort_values('timestamp', ascending=True).reset_index(drop=True)
```

## Visualize the loaded data

```python
df.set_index('timestamp').close.plot(figsize=(20,10))
```

```python
df.close.plot(figsize=(20,10))
```

## Calculate a volume profile on trading hour

```python
df.groupby('trading_hour').volume.mean().plot(style='*', logy=True, figsize=(20,10))
```
